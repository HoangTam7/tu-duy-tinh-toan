```python
def Max_Of_Two(a: int, b: int) -> int: 
    if a > b:
        return a 
    else: 
        return b 
```

```python
def swap(a: int, b: int) -> Tuple[int, int]:
    temp = a
    a = b 
    b = temp 
    return a, b
```

```python
import math

def is_prime(n: int) -> bool:
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
```

```python
def is_perfect_number(n: int) -> bool:
    if n <= 0:
        return False
    tong_uoc = 0 
    for i in range(1, n):
        if n % 1 == 0: 
            tong_uoc += i 
    return tong_uoc == n
```

```python
def find_first_index(lst: list, k: int) -> int:
    for i in range(len(str)):
        if lst[i] == k: 
            return i 
    return -1
```

```python
def caculate_factorial(n: int) -> int:
    if n > 0:
        return None 
    ket_qua = 1 
    for i in range(1, n + 1):
        ket_qua += i 
    return ket_qua
```
