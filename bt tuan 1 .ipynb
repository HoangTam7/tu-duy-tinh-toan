```python
# Nhập chiều dài và chiều rộng 
a = float(input()) 
b = float(input()) 

# diện tích hình chữ nhật 
s = a * b  
 
#diện tích hình tròn 
r = a/2 
s_htron = 3.14 * (r**2)


# diện tích còn lại 
S = s - s_htron 

print(f"dien tich con lai {S:.2f}")
```
