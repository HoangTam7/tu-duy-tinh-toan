```python
import math

a = float(input("Nhập cạnh a: "))
b = float(input("Nhập cạnh b: "))
c = float(input("Nhập cạnh c: "))

if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        loai = "Tam giác đều"
    elif a == b or b == c or a == c:
        if (a**2 + b**2 == round(c**2, 2)) or (a**2 + c**2 == round(b**2, 2)) or (b**2 + c**2 == round(a**2, 2)):
            loai = "Tam giác vuông cân"
        else:
            loai = "Tam giác cân"
    elif (a**2 + b**2 == round(c**2, 2)) or (a**2 + c**2 == round(b**2, 2)) or (b**2 + c**2 == round(a**2, 2)):
        loai = "Tam giác vuông"
    else:
        loai = "Tam giác thường"
    p_nua = (a + b + c) / 2
    dien_tich = math.sqrt(p_nua * (p_nua - a) * (p_nua - b) * (p_nua - c))
    
    print(f"Đây là: {loai}")
    print(f"Chu vi: {a + b + c}")
    print(f"Diện tích: {round(dien_tich, 2)}")
else:
    print("Khong la ba canh cua mot tam giac")
```

```python
import math

a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
c = float(input("Nhập c: "))

if a == 0:
    if b == 0:
        print("Vô số nghiệm" if c == 0 else "Vô nghiệm")
    else:
        print(f"Nghiem x = {-c/b}")
else:
    delta = b**2 - 4*a*c
    if delta < 0:
        print("Phương trình vô nghiệm")
    elif delta == 0:
        print(f"Phương trình có nghiệm kép x = {-b/(2*a)}")
    else:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        print(f"x1 = {x1}, x2 = {x2}")
```

```python
import math

x0 = float(input("Tâm x0: "))
y0 = float(input("Tâm y0: "))
r = float(input("Bán kính R: "))
xa = float(input("Điểm xA: "))
ya = float(input("Điểm yA: "))

khoang_cach = math.sqrt((xa - x0)**2 + (ya - y0)**2)

if round(khoang_cach, 5) == round(r, 5):
    print("Điểm A nằm TRÊN đường tròn")
elif khoang_cach < r:
    print("Điểm A nằm TRONG đường tròn")
else:
    print("Điểm A nằm NGOÀI đường tròn")
```

```python
import math

x = float(input("Nhập số thực x (cơ số): "))
y = float(input("Nhập số thực y (số mũ): "))

ket_qua_1 = x ** y

try:
    ket_qua_2 = math.pow(x, y)
    print(f"Kết quả {x}^{y} là: {ket_qua_2}")
except ValueError:
    print(f"Kết quả {x}^{y} là: {ket_qua_1}")
```
